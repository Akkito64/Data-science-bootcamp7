```python
# TODO 03 - your friend ask for `California` data, filter it and export csv for him

# Filter the dataset for California data
california_data = df.loc[df['State'] == 'California']

# Export the filtered data to a CSV file
california_data.to_csv('california_data.csv', index=False)
```

```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file
# Filter the dataset for California and Texas data in 2017
filtered_data = df.loc[(df['State'].isin(['California', 'Texas'])) & (df['Order Date'].dt.year == 2017)]

# Export the filtered data to a CSV file
filtered_data.to_csv('california_texas_2017_data.csv', index=False)


```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017

# Filter the dataset for 2017 data
data_2017 = df[df['Order Date'].dt.year == 2017]

# Calculate the total sales
total_sales_2017 = data_2017['Sales'].sum()

# Calculate the average sales
average_sales_2017 = data_2017['Sales'].mean()

# Calculate the standard deviation of sales
std_sales_2017 = data_2017['Sales'].std()

# Print the results
print("Total Sales in 2017:", round(total_sales_2017,3))
print("Average Sales in 2017:", round(average_sales_2017,3))
print("Standard Deviation of Sales in 2017:", round(std_sales_2017,3))


```

```python
# TODO 06 - which Segment has the highest profit in 2018

exam06 = df[df['Order Date'].dt.year == 2018].groupby('Segment')['Profit'].max()

display(exam06)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019

sales = df[(df['Order Date'] > '2019-04-15') & (df['Order Date'] < '2019-12-31')].groupby('State')['Sales'].sum().sort_values(ascending= True).head()

sales
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 

# Filter the dataset for 2019 data in the West and Central regions
data_2019_west_central = df[(df['Order Date'].dt.year == 2019) & (df['Region'].isin(['West', 'Central']))]

# Calculate the total sales for West and Central in 2019
sales_west_central_2019 = data_2019_west_central['Sales'].sum()

# Calculate the total sales for the entire dataset in 2019
total_sales_2019 = df[df['Order Date'].dt.year == 2019]['Sales'].sum()

# Calculate the proportion of sales in West and Central in 2019
proportion_sales_west_central_2019 = round((sales_west_central_2019 / total_sales_2019) * 100,2)

# Print the proportion of sales in West and Central in 2019
print("Proportion of total sales (%) in West + Central in 2019:", int(proportion_sales_west_central_2019), "%")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020

df_product = df[(df['Order Date'].dt.year >= 2019) & (df['Order Date'].dt.year <= 2020)].groupby("Product Name")['Product ID'].count().sort_values(ascending = False).head(10)

df_Total_Salse = df[(df['Order Date'].dt.year >= 2019) & (df['Order Date'].dt.year <= 2020)].groupby("Product Name")['Sales'].sum().sort_values(ascending = False).head(10)

print("top 10 popular products in terms of number of orders\n",df_product,"\n")

print("----------------------------------------------------------------------------------\n")

print("top 10 popular products in terms of total sales\n",df_Total_Salse,"\n")
```

```python
import matplotlib.pyplot as plt
# TODO 10 - plot at least 2 plots, any plot you think interesting :)

# plot 1 : 'Sales * Quantity for Each Ship Mode'

# Calculate the sales multiplied by quantity
df['Total'] = df['Sales'] * df['Quantity']

# Group the data by ship mode and calculate the sum of the 'Total' column
grouped_data = df.groupby('Ship Mode')['Total'].sum().sort_values(ascending=False)

# Plot the data
plt.subplot(1,1,1)
plt.bar(grouped_data.index, grouped_data.values)
plt.xlabel('Ship Mode')
plt.ylabel('Sales * Quantity')
plt.title('Sales * Quantity for Each Ship Mode')
plt.show()


#plot 2 'Sales * Quantity for each Category

# Group the data by category and calculate the sum of the 'Total' column
grouped_data = df.groupby('Category')['Total'].sum()

# Sort the data in descending order and select the top 10 categories
categories = grouped_data.nlargest(10)

# Plot the data
plt.subplot(2,1,1)
plt.bar(categories.index, categories.values)
plt.xlabel('Category')
plt.ylabel('Sales * Quantity')
plt.title('Categories by Sales * Quantity')
plt.xticks(rotation=90)
plt.show()
```
